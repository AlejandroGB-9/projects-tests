```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import pickle

#Abrimos competicion

comp_df = pd.read_csv("./comp_st9ns1.txt.bz2", compression= "bz2", index_col=0)
cs = comp_df.T

# Cojemos los datos de competicion
C=[]

for i in cs.columns:
    C.append(cs[i].tolist())

# We open the .pkl
with open('modelo_final.pkl', 'rb') as f:
    model_dict = pickle.load(f)

model_rmse = model_dict['model_rmse']
model_mae = model_dict['model_mae']


#--------------------------------------------------------------------------------------------------------------
# Predictions with comp.
#--------------------------------------------------------------------------------------------------------------
# Escalamos los datos de competición porque tal y como hemos visto escalarlos da mejores resultados.
scaler = StandardScaler()
C_train_scaled = scaler.fit_transform(C)

predicciones_RMSE = model_rmse.predict(C_train_scaled)
predicciones_MAE = model_mae.predict(C_train_scaled)

#Pasamos las predicciones a un .csv
pd.DataFrame({'predicciones_RMSE': predicciones_RMSE,'predicciones_MAE': predicciones_MAE}).to_csv('predicciones.csv', index=False)
```
